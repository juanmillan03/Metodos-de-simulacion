```python
import numpy as np

def circle_equation(x1, y1, x2, y2, x3, y3):
    # Crear matrices para el sistema de ecuaciones
    A = np.array([[x1 - x2, y1 - y2], [x1 - x3, y1 - y3]])
    b = np.array([(x1**2 - x2**2 + y1**2 - y2**2) / 2, (x1**2 - x3**2 + y1**2 - y3**2) / 2])

    # Resolver el sistema de ecuaciones
    h, k = np.linalg.solve(A, b)

    # Calcular el radio
    r = np.sqrt((x1 - h)**2 + (y1 - k)**2)

    return f"(x - {h})^2 + (y - {k})^2 = {r}^2"

x1, y1 = 64, 0
x2, y2 = 96, 64
x3, y3 = -64, 128

print(circle_equation(x1, y1, x2, y2, x3, y3))
```

```python
Dis_F = np.array((175, 165, 155, 145, 140, 135, 130, 128, 125, 120, 115)) - 96
n_ref = np.array((1.5, 1.6, 1.7, 1.8, 1.9, 2.0, 2.1, 2.2, 2.3, 2.4, 2.5))

ln_DisF = np.log(Dis_F)
ln_nref = np.log(n_ref)

# Ajustar línea recta
coeffs = np.polyfit(ln_DisF, ln_nref, 1)
poly = np.poly1d(coeffs)

m=coeffs[0]
b=coeffs[1]

# Graficar datos y línea recta ajustada
plt.plot(ln_DisF, ln_nref, "o", label="Datos")
plt.plot(ln_DisF, poly(ln_DisF), "-", label=r"$\ln(f)$= "f"{b:.2f}{m:.2f}"r"$\ln(n)$")
plt.xlabel("ln(Dis_F)")
plt.ylabel("ln(n_ref)")
plt.legend()
plt.show()

print('Ley de potencias-------> f=' f"{np.exp(m):.2f}+n^{-0.38}")




```
