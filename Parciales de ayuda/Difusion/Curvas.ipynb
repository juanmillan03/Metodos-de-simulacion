## PUNTO 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATOS2 = np.genfromtxt('Varianza.dat', delimiter=' ', usecols=(0, 1))
t= DATOS2[:,0]
v= DATOS2[:,1]

# Ajustar línea recta
coeffs = np.polyfit(t, v, 1)
poly = np.poly1d(coeffs)

m=coeffs[0]
b=coeffs[1]

# Graficar datos y línea recta ajustada
plt.plot(t, v, "o", label="Datos")
plt.plot(t, poly(t), "-", label=r"$\sigma^{2}$="f"{b:.2f}+{m:.2f}"r"$t$")
plt.ylabel(r"$\sigma^{2}$")
plt.xlabel("t")
plt.legend()
plt.show()

print(f'2D={m:.4f}')


```
