```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python

# Cargar datos
sigma = np.loadtxt('Resultados/sigma2.dat')
t = np.array(sigma[:, 0]).reshape(-1, 1)  # Primera columna
sx = np.array(sigma[:, 1]).reshape(-1, 1)  # Segunda columna
sy = np.array(sigma[:, 2]).reshape(-1, 1)  # Segunda columna

# Crear el modelo de regresión lineal
model_x = linear_model.LinearRegression()
model_x.fit(t, sx)
pendiente = model_x.coef_[0][0]
intercepto = model_x.intercept_[0]

# Calcular las predicciones
y_pred = model_x.predict(t)

# Calcular el error estándar de la pendiente
n = len(t)
residuos = sx - y_pred
residuos_var = np.var(residuos, ddof=1)
x_var = np.var(sx, ddof=1)
error_pendiente = np.sqrt(residuos_var / (n * x_var))

print(f"La pendiente de la regresión lineal es: {pendiente}")
print(f"El error estándar de la pendiente es: {error_pendiente}")

model_y = linear_model.LinearRegression()
model_y.fit(t, sy)
pendiente1 = model_y.coef_[0][0]
intercepto = model_y.intercept_[0]

# Calcular las predicciones
y_pred_y= model_y.predict(t)

# Calcular el error estándar de la pendiente
n = len(t)
residuos = sy - y_pred_y
residuos_var = np.var(residuos, ddof=1)
x_var = np.var(sy, ddof=1)
error_pendiente = np.sqrt(residuos_var / (n * x_var))

print(f"La pendiente de la regresión lineal es: {pendiente1}")
print(f"El error estándar de la pendiente es: {error_pendiente}")
print(f"La pendiente de la regresión lineal es: {pendiente1+pendiente}")

# Graficar los datos y la línea de regresión
plt.scatter(t, sx, s=2)
plt.scatter(t, sy, s=2)
plt.plot(t, y_pred, linewidth=2)
plt.plot(t, y_pred_y, linewidth=2)
plt.grid()
plt.show()
```

```python
fig, axs = plt.subplots(3, 3, figsize=(10, 10))
for i in range(3):
    for j in range(3):
        # Cargar los datos correspondientes al índice actual
        datos = np.loadtxt(f'Resultados/Cuadricula_{(i*3+j)*50}.dat')
        
        ax = sns.heatmap(datos, cmap='viridis', ax=axs[i, j])
        
        # Quitar los ejes
        ax.set(xticks=[], yticks=[])
        
        # Agregar título al subplot
        ax.set_title(f'Tmax={(i*3+j)*50}')

# Ajustar el espacio entre subplots
plt.tight_layout()

# Mostrar la figura completa con todos los subplots
plt.show()
```
