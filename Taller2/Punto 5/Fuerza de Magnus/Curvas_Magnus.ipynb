```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
FuerzaTot = np.genfromtxt('FuerzasOmega.dat', delimiter=' ', usecols=(0, 1))

# Realizar regresión lineal
z = np.polyfit(FuerzaTot[:,0], np.abs(FuerzaTot[:,1]), 1)
p = np.poly1d(z)

# Obtener pendiente y punto de corte
pendiente = z[0]
punto_corte = z[1]

print(f"Pendiente: {pendiente:.4f}")
print(f"Punto de corte: {punto_corte:.4f}")

# Plotear datos y regresión lineal
plt.plot(FuerzaTot[:,0], np.abs(FuerzaTot[:,1]), 'o', label='Datos')
plt.plot(FuerzaTot[:,0],p(FuerzaTot[:,0]), '-', label='Regresión lineal')
plt.xlabel(r'$\omega$')
plt.ylabel(r'$F_{M}$')
plt.grid(color='#c0c0c0', linewidth=1,alpha=0.5)
plt.minorticks_on()
plt.legend()
plt.show()


R=8
A1=np.pi*R*R
A2=2*np.pi*R
V=0.1

CM=2*pendiente/(A2*R*V)
print(f'Cm={CM:.2f}')
```
