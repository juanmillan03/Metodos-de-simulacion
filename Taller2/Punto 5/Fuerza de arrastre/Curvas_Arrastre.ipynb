```python
import numpy as np
import matplotlib.pyplot as plt
```

## PUNTO D COEFICIENTE DE ARRASTRE

```python
Fuerzas = np.genfromtxt('Fuerzas.dat', delimiter=' ', usecols=(0, 2))

plt.plot(Fuerzas[:,0],np.abs(Fuerzas[:,1]), 'o-')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$F_{y}$')
plt.grid(color='#c0c0c0', linewidth=1,alpha=0.5)
plt.minorticks_on()
plt.legend()
plt.show()
```

```python
C_a = np.genfromtxt('CoefArrastre.dat', delimiter=' ', usecols=(0, 1))

plt.plot(C_a[:,0],C_a[:,1], 'o-')
plt.xlabel(r'$R_{e}$')
plt.ylabel(r'$C_{A}$')
plt.grid(color='#c0c0c0', linewidth=1,alpha=0.5)
plt.minorticks_on()
plt.legend()
plt.show()
```

```python
F_a = np.genfromtxt('FuerzaArrastre.dat', delimiter=' ', usecols=(0, 1))

plt.plot(F_a[:,0],np.abs(F_a[:,1]), 'o-')
plt.xlabel(r'$R_{e}$')
plt.ylabel(r'$F_{A}$')
plt.grid(color='#c0c0c0', linewidth=1,alpha=0.5)
plt.minorticks_on()
plt.legend()
plt.show()
```

```python
F_V = np.genfromtxt('FuerzaVec.dat', delimiter=' ')

# Separar las coordenadas x y y las magnitudes del vector
x = F_V[:, 0]
y = F_V[:, 1]
ux = F_V[:, 2]
uy = F_V[:, 3]

# Graficar el campo vectorial con quiver
plt.quiver(x, y, ux, uy, color='Green', label='Vectores de Fuerza')
plt.xlabel(r'$ix$')
plt.ylabel(r'$iy$')
plt.grid(color='#c0c0c0', linewidth=1,alpha=0.5)
plt.minorticks_on()
plt.legend()
plt.show()
```

```python
A_V = np.genfromtxt('VecArea.dat', delimiter=' ')

# Separar las coordenadas x y y las magnitudes del vector
Ax = A_V[:, 0]
Ay = A_V[:, 1]
Aux = A_V[:, 2]
Auy = A_V[:, 3]

# Graficar el campo vectorial con quiver
plt.quiver(Ax, Ay, Aux, Auy, color='Orange', label='Vectores de Area')
plt.xlabel(r'$ix$')
plt.ylabel(r'$iy$')
plt.grid(color='#c0c0c0', linewidth=1,alpha=0.5)
plt.minorticks_on()
plt.legend()
plt.show()

plt.plot(Ax, Ay,'o-', color='black')
plt.xlabel(r'$ix$')
plt.ylabel(r'$iy$')
plt.grid(color='#c0c0c0', linewidth=1,alpha=0.5)
plt.minorticks_on()
plt.legend()
plt.show()
```
