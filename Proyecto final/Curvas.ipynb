```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Perfil_velocidades = np.genfromtxt('Ondas3D.dat', delimiter=' ', usecols=(0, 1))
Perfil_velocidades2 = np.genfromtxt('Ondas3DSegunda.dat', delimiter=' ', usecols=(0, 1))

plt.plot(Perfil_velocidades[:,0],Perfil_velocidades[:,1],'o-', color='Black', alpha=0.5)
#plt.plot(Perfil_velocidades2[:,0],Perfil_velocidades2[:,1],'o-', color='Red', alpha=0.5)
plt.xlabel(r'$x$')
plt.ylabel(r'$p$')
plt.grid(color='#c0c0c0', linewidth=1,alpha=0.5)
plt.minorticks_on()
plt.show()
```
