```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import scipy as sci
import pandas as pd
```

```python
datos_1 = pd.read_csv('One_lennard.txt', delimiter=' ', header=None)

t_1 = datos_1.iloc[1:, 0].values
x_1 = datos_1.iloc[1:, 1].values

plt.plot(t_1, x_1, color='blue', label='Datos')
plt.xlabel('Tiempo')
plt.ylabel('Posicin')
plt.title('x vs t para una partcula en un potencial de Lennard-Jones')
#plt.legend()
plt.grid(True)
plt.show()
```

```python
datos_2 = pd.read_csv('Rgiro_vs_t_gas.txt', delimiter=' ', header=None)

R_gas = datos_2.iloc[1:, 0].values
t_2 = datos_2.iloc[1:, 1].values

plt.plot(t_2, R_gas, color='blue', label='Datos')
plt.xlabel('Tiempo')
plt.ylabel(r'$R_{gas}$')
plt.title('x vs t para una partcula en un potencial de Lennard-Jones')
#plt.legend()
plt.grid(True)
plt.show()
```
