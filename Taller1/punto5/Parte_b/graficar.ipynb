```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("mu_experimental.dat",sep=" ",names=["Lx","mu","v"])

plt.scatter(data["Lx"],data["mu"],s=20)
plt.xlabel("Lx")
plt.ylabel(r"$\mu$=y/x")
plt.grid()
plt.title(r"Resultados de $\mu$ respecto a longitud de la caja")
plt.savefig("Mu.pdf")
   
```
