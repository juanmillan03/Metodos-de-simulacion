```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("mu_experimental.dat",sep=" ",names=["Lx","mu"])

plt.scatter(data["Lx"],data["mu"],s=20)
plt.xlabel("Lx")
plt.ylabel(r"mu=y/x")
plt.grid()
plt.title("Resultados de mu")
plt.savefig("Mu.pdf")
   
```
